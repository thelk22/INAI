Two important aspects of matplotlib are figures and axes. Figures correspond to a
physical figure, while an axes corresponds to regions or sections of the figure. Therefore,
naturally, you first create a figure and add axes too it.

```python
%matplotlib inline
import matplotlib.pyplot as plt

fig, (ax1, ax2) = plt.subplots(2) # two plots..
fig, (ax1, ax2) = plt.subplots(2, sharex=True, sharey=True) # same as above, but with a shared x and y axis between the two plots
```

```python
import numpy as np

fig, (ax1, ax2, ax3) = plt.subplots(3)
sigma = 3
mu = 2
n = sigma * np.random.randn(500,1) + mu
ax1.plot(n)

mu_noise, sigma_noise = 1, .7
noise = sigma_noise * np.random.randn(500,1) + mu_noise
y = n * noise
ax2.scatter(n, y)
ax3.hist(n)

plt.tight_layout()
```

```python
# Generate RVs

n5000 = np.random.normal(1,1,5000)
plt.hist(n5000, bins=100)
```

```python
p3 = n5000 + 3
p4 = n5000 * 0.5 + 3

fig, (ax1, ax2) = plt.subplots(2)

ax1.hist(p3)
ax2.hist(p4)

print(p3.mean())
print(p3.var())
print(p3.var()**0.5)
```

```python
import pandas as pd

df = pd.read_csv('https://seis.bris.ac.uk/~rm17770/data1414.txt', sep='\s', header=None)
```

```python
print(df.head())
```

```python
labels = df.iloc[:,-1] 
data = df.iloc[:,:-1] 

plt.imshow(np.array(data.iloc[10]).reshape(14,14), cmap='gray')
```

```python
print(len(data))
print(data.mean())
print(data.max())
```

```python
plt.imshow(np.array(np.mean(data, axis=0)).reshape(14,14), cmap='gray')
```

```python
zero_idx = df.index[df[df.columns[-1]] == 0]
zero = data.loc[zero_idx]
```

```python
plt.imshow(np.array(np.mean(zero, axis=0)).reshape(14,14), cmap='gray')
```

```python
ones_idx = df.index[df[df.columns[-1]] == 1]
ones = data.loc[ones_idx]
plt.imshow(np.array(np.mean(ones, axis=0)).reshape(14,14), cmap='gray')
```

```python
pixels=[49,50,60,102]

for pixel in pixels:
    fig, (ax1, ax2) = plt.subplots(2, sharex=True, sharey=True)
    ax1.set_title('Zero Pixel '+str(pixel))
    ax1.hist(zero.iloc[:,pixel], bins=20)
    ax2.set_title('One Pixel '+str(pixel))
    ax2.hist(ones.iloc[:,pixel], bins=20)
    plt.show()
```
